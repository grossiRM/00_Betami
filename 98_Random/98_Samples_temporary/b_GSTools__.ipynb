```python
x = y = np.arange(100) ; model = gs.Gaussian(dim=2, var=1, len_scale=10)  ; srf = gs.SRF(model)
srf.set_pos([x, y], "structured")                                         ; ens_no = 4
for i in range(ens_no):
    srf(seed=i, store=f"field{i}")
fig, ax = pt.subplots(1, 4, sharex=True, sharey=True)   ; ax = ax.flatten()
for i in range(ens_no):
    ax[i].imshow(srf[i].T, origin="lower")
pt.show()
```

```python
x = y = np.arange(100) ; model = gs.Gaussian(dim=2, var=1, len_scale=10)  ; srf = gs.SRF(model)
srf.set_pos([x, y], "structured")                                         ; ens_no = 4
seed = MasterRNG(201705)
for i in range(ens_no):
    srf(seed=seed(), store=f"better_field{i}")
fig, ax = pt.subplots(1, 4, sharex=True, sharey=True)   ; ax = ax.flatten()
for i in range(ens_no):
    ax[i].imshow(srf[i].T, origin="lower")
pt.show()
```
