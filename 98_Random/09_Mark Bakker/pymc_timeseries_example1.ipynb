In this Notebook we perform MCMC sampling on a timeseries model of groundwater dynamics caused by daily variations in recharge.

###Generation of synthetic head data
Synthetic daily head data is used in combination with 'measured' recharge, which is approximated as measured precipitation minus measured potential evaporation; 1500 daily values of recharge are stored in the file `recharge.txt`. Heads are generated through convolution of the recharge data with a response function, plus a constant and a correlated residual:

$h(t) = d + \int_{-\infty}^t N(\tau)A\text{e}^{(t-\tau)/\alpha}\text{d}\tau + \varepsilon(t)$

where $d$ is an additive constant, $A$ and $a$ are parameters of the response function, and $\varepsilon(t)$ are correlated residuals defined here with an AR(1) model

$\varepsilon(t+\Delta t) = \alpha\varepsilon(t) + \nu(t+\delta t)$

where $\alpha$ is an autoregressive parameter, and $\nu$ is white noise (also called innovations) distributed as $\text{Normal}(0,\sigma)$ where $\sigma$ is the standard deviation. 

The following values are used for the response function: $A=6$ and $a=70$.

```python
Ao = 6.0
ao = 70.0
tmax = 500 # time when response function equals zero (fixed here for simplicity)
t = arange(tmax) # time in days
F = Ao * ao * (1-exp(-t/ao)) # step function
H = F[1:] - F[:-1] # one-day block function
subplot(211)
plot(t,F)
xlabel('time')
ylabel('step function')
title('true response function')
subplot(212)
plot(t[1:],H)
xlabel('time')
ylabel('block function')
```

This response function is convolved with the recharge data, the first 500 days are neglectd as warm-up (so we are left with 1000 head values), and the constant $d$ is added. This is the head variation without any introduced error.

```python
N = loadtxt('recharge.txt')
tstart = 500
d = 27.0
ho = convolve(N,H) + do
ho = ho[tstart:len(N)]
plot(ho)
xlabel('time')
ylabel('head (m)')
```

AR(1) correlated errors as defined above may be generated though convolution of randomly sampled noise (from the specified Normal distribution) and convolution with a response function $f(t_n)=\alpha^n$, $n=0,1,...$ The Normally distributed noise is generated with the `rnormal` function of `pymc`. The following values are used for the correlated residuals: $\alpha=0.95$, and $\sigma=0.04$. 

```python
import pymc
alphao = 0.95
sigo = 0.04
vo = pymc.rnormal(0,1.0/sigo**2,size=len(N)) # noise
Hv = alphao**arange(tmax) # response function
ro = convolve(vo,Hv) # correlated residuals
ro = ro[tstart:len(N)] 
subplot(211)
plot(vo[tstart:])
xlabel('time')
ylabel('noise')
subplot(212)
plot(ro)
xlabel('time')
ylabel('correlated residuals')
```

The correlated residuals are added to the generated heads to arrive at the synthetic head data that will be used in the analysis. This head data will be referred to as the observed heads.

```python
subplot(211)
plot(ho)
xlabel('time')
ylabel('head without errors')
ho += ro
subplot(212)
plot(ho)
xlabel('time')
ylabel('observed heads')
```

###Model construction
There are five parameters in the model: $d$, $A$, $a$, $\alpha$, and $\sigma$. First, we define the priors (all uninformative) as `Uniform` function, which requires specification of a name (note sure why that is needed), the minimum value, maximum value, and starting value

```python
A = pymc.Uniform('A',0.0,100.0,value=10.0)
a = pymc.Uniform('a', 0, 600.0, value= 50.0)
d = pymc.Uniform('d', 0.0, 100.0, value= ho.mean() )
alpha = pymc.Uniform('alpha', 0.0, 1.0, value=0.5)
sig = pymc.Uniform('sig', 1e-9, 1.0, value= 0.1)
```

The `head_model` function is a deterministic function of all the input variables and computes the head given the current values of the parameters

```python
@pymc.deterministic(plot=False)
def head_model(A=A, a=a, d=d, alpha=alpha, N=N, ho=ho):
    Nspinup = 500
    t = arange(500)            # response function is zero at t=500 
    F = A * a * (1-exp(-t/a))  # step function
    H = F[1:] - F[:-1]         # block function
    h = d + convolve(N,H)      # convolution
    h = h[Nspinup:len(N)]      # remove spin-up
    h[1:] += alpha * (ho[0:-1] - h[0:-1])  # AR(1) residuals
    return h
```

The loglikelihood function has a mean equal to `head_model` and a precision equal to $1/\sigma^2$. The values are the observed heads `ho`, and the keyword `observed` is set to `True`.

```python
loglike = pymc.Normal('l', mu = head_model, tau=1.0/sig**2, value = ho, observed=True) 
mcmc = pymc.MCMC([d, A, a, alpha, sig, loglike]) 
```

Ten thousand samples are drawn. This is not very many yet, but is sufficient here to show how things work.

```python
mcmc.sample(10000)
```

Traces of the samples are plotted. It appears that the first 5000 values need to be discarded as burn in.
