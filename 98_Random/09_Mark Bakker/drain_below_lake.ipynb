```python
%pylab inline
rcParams['contour.negative_linestyle'] = 'Solid'
```

```python
# Input values
L = 20.0
H = 20.0
Qleft = -1.0  # positive value is outflow at left
Qright = 0.5 # positive value is outflow at right
xd = 0
yd = -H/2
Qd = 0.5  # positive value is outflow of drain
# Computed values
lab = exp(-pi * L / (2*H)) - 1e-6*1j
one = 1.0 - 1e-6*1j
zd = xd + yd * 1j
```

```python
def zetafunc(z):
    rv = sqrt( (1-exp(-pi*z/H)) / (1-exp(-pi*(z-L)/H)) )
    return rv
```

```python
def omega(x,y):
    z = x + y*1j
    zeta = zetafunc(z)
    zeta1 = zetafunc(zd)
    zeta2 = -zeta1.real + zeta1.imag * 1j
    zeta3 = -zeta1.real - zeta1.imag * 1j
    zeta4 = zeta1.real - zeta1.imag * 1j
    om = Qleft / pi * log((zeta - lab)/(zeta + lab)) + Qright / pi * log((zeta - one)/(zeta + one)) + \
         Qd / (2*pi) * log( ((zeta-zeta1)*(zeta-zeta4)) / ((zeta-zeta2)*(zeta-zeta3)) )
    return om
```

```python
x,y = meshgrid(linspace(-L,2*L,100),linspace(-H,0,100))
om = omega(x,y)
```

```python
contour(x,y,om.real,arange(-10,10,0.1),colors='g')
contour(x,y,om.imag,arange(-2,2.001,0.1),colors='b')
axis('scaled')
print nanmin(om.real),nanmax(om.real)
print nanmin(om.imag),nanmax(om.imag)
```
