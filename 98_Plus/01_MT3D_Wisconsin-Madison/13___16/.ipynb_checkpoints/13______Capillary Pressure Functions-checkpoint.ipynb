### **van Genuchten, Brooks Corey**

```python
Swr = 0.1     ; Snwr = 0                                         #  residual water saturation | residual gas saturation (after imbibition)
Sw = np.linspace((Swr + 0.001), (1 - Snwr - 0.001), num=100)     # water saturation
Pc_entry = 2  ; m = 2; n = 3                                     # capillary entry pressure (kPa) in selected units
m_bc = 3      ; Pc_entry_bc = 0.9    
```

```python
def van_g_pc(Sw, Swr, Snwr, Pc_entry, m, n):
    Se = (Sw - Swr)/((1 - Snwr) - Swr)                ;  Pc = Pc_entry*(Se**(-1/m)-1)**(1/n)
    return Pc                                                                                     #    effective saturation
def brooks_corey_pc(Sw, Swr, Snwr, Pc_entry, m):
    Se = (Sw - Swr)/((1 - Snwr) - Swr)                ;  Pc = Pc_entry*(Se**(-1/m))
    return Pc
```

```python
Pc_vg  = van_g_pc        (Sw, Swr, Snwr, Pc_entry    , m, n)
Pc_vg2 = van_g_pc        (Sw, Swr, Snwr, Pc_entry*3  , m, n)
Pc_bc  = brooks_corey_pc (Sw, Swr, Snwr, Pc_entry_bc , m_bc)

plt.figure(figsize=(6, 2), dpi=100)          ; plt.plot  (Sw, Pc_vg)   ; plt.plot(Sw, Pc_vg2)   
plt.xlabel('Water Saturation', fontsize=10)  ; plt.ylabel('Cap. Pressure (kPa)', fontsize=10)  ; plt.show()
plt.figure(figsize=(6, 2), dpi=100)          ; plt.plot  (Sw, Pc_vg)   ; plt.plot(Sw, Pc_bc )
plt.xlabel('Water Saturation')               ; plt.ylabel('Cap. Pressure (kPa)')          ; plt.show()
```
