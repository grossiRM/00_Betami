## **MNW2 : backward tracking simulation using particles released at MNW well.**

```python
pthobj.write_shapefile(well_pathlines ,direction="starting"   ,shpname=os.path.join(model_ws_m ,"pathlines.shp")      ,mg=m.modelgrid ,verbose=False);
pthobj.write_shapefile(well_pathlines ,one_per_particle=False ,shpname=os.path.join(model_ws_m ,"pathlines_1per.shp") ,mg=m.modelgrid ,verbose=False);

m2 = flopy.modflow.Modflow.load("EXAMPLE.nam" ,model_ws=str(model_ws_m) ,exe_name="mf2005"); 
print('t,z,x,y',m2.nrow_ncol_nlay_nper, '        wel.stress_period', m2.wel.stress_period_data.data)
node_data=np.array([(3,12,12,"well1","skin",-1,0,0,0,1.0,2.0,5.0,6.2),(4,12,12,"well1","skin",-1,0,0,0,0.5,2.0,5.0,6.2),],
                   dtype=[("k",int),("i",int),("j",int),("wellid",object),("losstype",object),("pumploc",int),("qlimit",int),
                          ("ppflag",int),("pumpcap",int),("rw",float),("rskin",float),("kskin",float),("zpump",float),],).view(np.recarray)
stress_period_data={0:np.array([(0,"well1",-150000.0)],dtype=[("per",int),("wellid",object),("qdes",float)])}
m2.get_package_list()
```

```python
m2.name = "Example_mnw"  ; m2.remove_package("WEL")
mnw2 = flopy.modflow.ModflowMnw2( model=m2,  mnwmax=1,node_data=node_data,stress_period_data=stress_period_data,itmp=[1, -1, -1])
m2.get_package_list()
```

```python
m2.change_model_ws(sim_ws_m2)
m2.write_name_file() ; m2.write_input()
success, buff = m2.run_model(silent=True, report=True)  ; assert success, pformat(buff)
```

```python
mp = flopy.modpath.Modpath6(modelname="ex6mnw",exe_name="mp6",modflowmodel =m2,  model_ws = model_ws_m2)
mpb = flopy.modpath.Modpath6Bas(mp, hdry=m2.lpf.hdry, laytyp=m2.lpf.laytyp, ibound=1, prsity=0.1)
sim = mp.create_mpsim(trackdir="backward", simtype="pathline", packages="MNW2")
```

```python
mp.change_model_ws(model_ws_m2)
mp.write_name_file()
mp.write_input()
success, buff = mp.run_model(silent=True, report=True)
if success:
    for line in buff:
        print(line)
else:
    raise ValueError("Failed to run.")
```
