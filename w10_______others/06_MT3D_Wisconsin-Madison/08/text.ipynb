```python
zoo_data = pd.read_csv("Zoo/zoo_data.csv", encoding = 'utf-8', index_col = ["animal_name"])   # print(zoo_data.head())
```

```python
pca = PCA(3)
pca.fit(zoo_data)
pca_data = pd.DataFrame(pca.transform(zoo_data))
print(pca_data.head())
```

```python
clusters = 7   ; kmeans = KMeans(n_clusters = clusters)    ; kmeans.fit(zoo_data)   ; print(kmeans.labels_)
```
