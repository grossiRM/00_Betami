```python
import numpy as np ; import matplotlib as mpl ; import matplotlib.pyplot as plt ; import pandas as pd
import flopy ; from flopy.plot import styles;  from pathlib import Path; import os 
```

```python
sim_ws1    = Path("../bin/Flopy_data/mf6-freyberg")  ; sim_ws2 = Path("./03a")  ; sim_ws2.mkdir(exist_ok=True)   
exe_name = 'E:\\15_REPOS\\00_BETAMI\\bin\\mf6.exe'
```

```python
sim = flopy.mf6.MFSimulation.load(sim_ws = sim_ws1, verbosity_level=0, exe_name=exe_name)
gwf = sim.get_model("freyberg")  ; modelgrid = gwf.modelgrid  
modelgrid.set_coord_info(xoff=0,yoff=0,angrot=0,crs="+proj=utm +zone=14 +ellps=WGS84 +datum=WGS84 +units=m +no_defs")
print(f"Grid shape: {modelgrid.shape}\n") 
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(8, 3), subplot_kw={"aspect": "equal"})
arrays = [modelgrid.top, modelgrid.botm, modelgrid.cell_thickness]  ; labels = ["top", "botm", "thickness"]
for ix, ax in enumerate(axs):
    pmv = flopy.plot.PlotMapView(modelgrid = modelgrid, ax=ax) ; pc = pmv.plot_array(arrays[ix], masked_values=[1e30], vmin=0, vmax=35, alpha=0.5)
    pmv.plot_grid(lw=0.5) ; pmv.plot_inactive() ; ax.set_title(f"{labels[ix]}", fontsize=10)
plt.colorbar(pc)
```

```python
x = modelgrid.xoffset + np.random.rand(10)*4000 ; y=modelgrid.yoffset + np.random.rand(10) * 10000  ;q=np.random.rand(10) * -0.02 ; pdata = []
for ix, xc in enumerate(x):
    i, j = modelgrid.intersect(xc, y[ix], forgive=True)
    if not np.isnan(i) and not np.isnan(j):
        if modelgrid.idomain[0, i, j]:  pdata.append([(0, i, j), q[ix]])
gwf.remove_package("wel_0")  ; stress_period_data = {0: pdata}  ; wel = flopy.mf6.modflow.ModflowGwfwel(gwf, stress_period_data=stress_period_data)
```

```python
fig, ax = plt.subplots(figsize=(8, 3), subplot_kw={"aspect": "equal"})
pmv = flopy.plot.PlotMapView(modelgrid=modelgrid, ax=ax) ; pmv.plot_grid(lw=0.5) ; pmv.plot_inactive() ; pmv.plot_bc(package=wel)
```
